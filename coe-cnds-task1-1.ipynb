```python
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
import cv2
import numpy as np
import gc  # Garbage collection

def preprocess_image(image):
    """Preprocess scanned images to improve OCR accuracy."""
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
    
    # Apply thresholding to remove background noise
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    
    # Apply dilation to enhance text
    kernel = np.ones((1, 1), np.uint8)
    processed_image = cv2.dilate(binary, kernel, iterations=1)
    
    return Image.fromarray(processed_image)

def extract_text_from_pdf(pdf_path, batch_size=50):
    """Extract text from scanned PDFs using OCR in batches for memory optimization."""
    try:
        text = ''
        
        # Convert PDF to images
        images = convert_from_path(pdf_path, dpi=300) 
        
        # Process images in batches
        for i in range(0, len(images), batch_size):
            batch_images = images[i:i + batch_size]  
            
            for image in batch_images:
                processed_image = preprocess_image(image)
                text += pytesseract.image_to_string(processed_image, 
                                                    lang='mar+eng+ben+chi+urd',  
                                                    config="--psm 6 --oem 1")  
            
            # Free memory after processing the batch
            del batch_images
            gc.collect()
        
        return text
    except Exception as e:
        print(f"Error extracting text: {e}")
        return None


pdf_path = "/kaggle/input/pdddffs/allpdfs/আহম্মেদNOC.pdf"
extracted_text = extract_text_from_pdf(pdf_path, batch_size=50)
print(extracted_text)
```

```python
from langdetect import detect_langs

def detect_language_unicode(text):
    """Detects language probabilities based on Unicode character distribution."""
    
    
    lang_counts = {
        'bengali': sum(1 for char in text if '\u0980' <= char <= '\u09FF'),
        'chinese': sum(1 for char in text if '\u4E00' <= char <= '\u9FFF'),
        'urdu': sum(1 for char in text if '\u0600' <= char <= '\u06FF'),
        'marathi': sum(1 for char in text if '\u0900' <= char <= '\u097F'),
        'english': sum(1 for char in text if '\u0041' <= char <= '\u007A'),
    }
    
    total_chars = sum(lang_counts.values())  

 
    lang_probs = {lang: count / total_chars for lang, count in lang_counts.items() if total_chars > 0}

    max_prob = max(lang_probs.values(), default=0)
    majority_languages = [lang for lang, prob in lang_probs.items() if prob >= max_prob * 0.9]  # Keep close to dominant

    cleaned_text = "".join(char for char in text if any(
        (lang == 'bengali' and '\u0980' <= char <= '\u09FF') or
        (lang == 'chinese' and '\u4E00' <= char <= '\u9FFF') or
        (lang == 'urdu' and '\u0600' <= char <= '\u06FF') or
        (lang == 'marathi' and '\u0900' <= char <= '\u097F') or
        (lang == 'english' and '\u0041' <= char <= '\u007A')
        for lang in majority_languages
    ))

    return majority_languages, lang_probs, cleaned_text  

majority_langs, probs, filtered_text = detect_language_unicode(extracted_text)

print("Detected Language Probabilities:", probs)
print("Majority Language(s):", majority_langs)
print("Filtered Text:", filtered_text)
```
