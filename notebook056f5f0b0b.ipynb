```python

import os
import zipfile
import pytesseract
import pdf2image
import cv2
from pdfminer.high_level import extract_text


ZIP_FILE_PATH = "/kaggle/input/pdddffs/allpdfs"  
EXTRACTED_FOLDER = "/kaggle/working/extracted_pdfs/"
OUTPUT_FOLDER = "/kaggle/working/output_texts/"

OCR_LANGUAGES = {"mar": "Marathi", "urd": "Urdu", "ben": "Bengali", "chi_sim": "Chinese", "eng": "English"}




def extract_text_from_pdf(pdf_path):
    try:
        text = extract_text(pdf_path).strip()
        if text:
            return text
    except Exception as e:
        print(f"Error extracting selectable text: {e}")
    return None


def extract_text_from_image(pdf_path):
    images = pdf2image.convert_from_path(pdf_path)
    extracted_text = ""
    
    for img in images:
        for lang_code in OCR_LANGUAGES.keys():
            text = pytesseract.image_to_string(img, lang=lang_code)
            if text.strip():
                extracted_text += f"\n[{OCR_LANGUAGES[lang_code]}]\n{text.strip()}\n"
    
    return extracted_text.strip()


def process_pdfs(pdf_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)  
    
    for pdf_file in os.listdir(pdf_folder):
        if pdf_file.endswith(".pdf"):
            pdf_path = os.path.join(pdf_folder, pdf_file)
            print(f"Processing: {pdf_file}")

            # First attempt direct text extraction
            text = extract_text_from_pdf(pdf_path)

            # If no text found, try OCR for scanned PDFs
            if not text:
                text = extract_text_from_image(pdf_path)

            # Save extracted text
            if text:
                output_txt_path = os.path.join(output_folder, pdf_file.replace(".pdf", ".txt"))
                with open(output_txt_path, "w", encoding="utf-8") as f:
                    f.write(text)
                print(f"Saved extracted text to: {output_txt_path}")
            else:
                print(f"No text found in: {pdf_file}")



   
    process_pdfs(EXTRACTED_FOLDER, OUTPUT_FOLDER)

 
```
