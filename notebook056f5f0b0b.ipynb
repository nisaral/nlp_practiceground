```python

import os
import zipfile
import pytesseract
import pdf2image
import cv2
from pdfminer.high_level import extract_text


ZIP_FILE_PATH = "/kaggle/input/pdddffs/allpdfs"  
EXTRACTED_FOLDER = "/kaggle/working/extracted_pdfs/"
OUTPUT_FOLDER = "/kaggle/working/output_texts/"

OCR_LANGUAGES = {"mar": "Marathi", "urd": "Urdu", "ben": "Bengali", "chi_sim": "Chinese", "eng": "English"}




def extract_text_from_pdf(pdf_path):
    try:
        text = extract_text(pdf_path).strip()
        if text:
            return text
    except Exception as e:
        print(f"Error extracting selectable text: {e}")
    return None


def extract_text_from_image(pdf_path):
    images = pdf2image.convert_from_path(pdf_path)
    extracted_text = ""
    
    for img in images:
        for lang_code in OCR_LANGUAGES.keys():
            text = pytesseract.image_to_string(img, lang=lang_code)
            if text.strip():
                extracted_text += f"\n[{OCR_LANGUAGES[lang_code]}]\n{text.strip()}\n"
    
    return extracted_text.strip()


def process_pdfs(pdf_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)  
    
    for pdf_file in os.listdir(pdf_folder):
        if pdf_file.endswith(".pdf"):
            pdf_path = os.path.join(pdf_folder, pdf_file)
            print(f"Processing: {pdf_file}")

            # First attempt direct text extraction
            text = extract_text_from_pdf(pdf_path)

            # If no text found, try OCR for scanned PDFs
            if not text:
                text = extract_text_from_image(pdf_path)

            # Save extracted text
            if text:
                output_txt_path = os.path.join(output_folder, pdf_file.replace(".pdf", ".txt"))
                with open(output_txt_path, "w", encoding="utf-8") as f:
                    f.write(text)
                print(f"Saved extracted text to: {output_txt_path}")
            else:
                print(f"No text found in: {pdf_file}")



   
    process_pdfs(EXTRACTED_FOLDER, OUTPUT_FOLDER)

 
```

```python
import os
import pytesseract
import pdf2image
import cv2
import numpy as np
from pdfminer.high_level import extract_text
from paddleocr import PaddleOCR


ocr_model = PaddleOCR(use_angle_cls=True, lang="en")  


EXTRACTED_FOLDER = "/kaggle/input/pdddffs/allpdfs/"  


OUTPUT_FOLDER = "/kaggle/working/output_texts/"
os.makedirs(OUTPUT_FOLDER, exist_ok=True) 


pdf_files = [os.path.join(EXTRACTED_FOLDER, f) for f in os.listdir(EXTRACTED_FOLDER) if f.endswith(".pdf")]

def detect_pdf_type(pdf_path):
    """Determine if a PDF contains selectable text or scanned images."""
    text = extract_text(pdf_path)
    if text.strip():  
        return "selectable_text"

    images = pdf2image.convert_from_path(pdf_path, dpi=300)
    sample_img = np.array(images[0])

    gray = cv2.cvtColor(sample_img, cv2.COLOR_RGB2GRAY)
    std_dev = np.std(gray)

    return "bw_scanned" if std_dev < 30 else "colorful_scanned"

def extract_text_from_pdf(pdf_path, pdf_type, lang):
    """Extract text from a PDF based on its type."""
    extracted_text = ""

    if pdf_type == "selectable_text":
        extracted_text = extract_text(pdf_path)
    
    elif pdf_type == "colorful_scanned":
        images = pdf2image.convert_from_path(pdf_path, dpi=300)
        for img in images:
            result = ocr_model.ocr(np.array(img), cls=True)
            extracted_text += "\n".join([line[1][0] for line in result[0] if line[1][0]]) + "\n"

    elif pdf_type == "bw_scanned":
        images = pdf2image.convert_from_path(pdf_path, dpi=300)
        for img in images:
            extracted_text += pytesseract.image_to_string(img, lang=lang) + "\n"

    return extracted_text


for pdf_file in pdf_files:
    print(f"Processing: {pdf_file}")

    pdf_type = detect_pdf_type(pdf_file)

    
    filename_lower = pdf_file.lower()
    if "marathi" in filename_lower:
        lang = "mar"
    elif "bengali" in filename_lower:
        lang = "ben"
    elif "urdu" in filename_lower:
        lang = "urd"
    elif "chinese" in filename_lower:
        lang = "chi_sim"
    else:
        lang = "eng"

    
    extracted_text = extract_text_from_pdf(pdf_file, pdf_type, lang)

    
    output_txt_path = os.path.join(OUTPUT_FOLDER, os.path.basename(pdf_file).replace(".pdf", ".txt"))
    with open(output_txt_path, "w", encoding="utf-8") as f:
        f.write(extracted_text)

print("✅ Text extraction completed. Check the /kaggle/working/output_texts/ folder.")
```

```python
import os
from langdetect import detect, DetectorFactory


DetectorFactory.seed = 0


EXTRACTED_TEXT_FOLDER = "/kaggle/input/smaplee/sample_txt"
LANGUAGE_TEXT_FOLDER = "/kaggle/working/language_wise_texts/"

os.makedirs(LANGUAGE_TEXT_FOLDER, exist_ok=True)


LANG_MAP = {
    "mr": "Marathi",
    #"ur": "Urdu",
    #"bn": "Bengali",
    "zh-cn": "Chinese",
    "en": "English"
}

def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"

for text_file in os.listdir(EXTRACTED_TEXT_FOLDER):
    if text_file.endswith(".txt"):
        text_file_path = os.path.join(EXTRACTED_TEXT_FOLDER, text_file)

        with open(text_file_path, "r", encoding="utf-8") as file:
            text = file.read()

        lines = text.split("\n")

        lang_texts = {key: [] for key in LANG_MAP.keys()}

        for line in lines:
            lang_code = detect_language(line)
            if lang_code in lang_texts:
                lang_texts[lang_code].append(line)

        for lang_code, content in lang_texts.items():
            if content:  
                lang_folder = os.path.join(LANGUAGE_TEXT_FOLDER, LANG_MAP[lang_code])
                os.makedirs(lang_folder, exist_ok=True)

                output_file = os.path.join(lang_folder, text_file)
                with open(output_file, "w", encoding="utf-8") as f:
                    f.write("\n".join(content))

print("Text classification and language-wise saving completed!")
```
